```python
from keras.datasets import imdb
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)
```

```python
x_train = pad_sequences(x_train, padding='post', maxlen=50)
x_test = pad_sequences(x_test, padding='post', maxlen=50)
```

```python
model = Sequential()
model.add(Embedding(input_dim=10000,output_dim=2, input_length=50))
model.add(SimpleRNN(32, return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentrop')
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
# Get the word index used by the IMDB dataset
word_index = imdb.get_word_index()

adjusted_word_index = {word: (index + 3) for word, index in word_index.items()}

# Add special tokens
special_tokens = {
    "<PAD>": 0,
    "<START>": 1,
    "<UNK>": 2,
    "<UNUSED>": 3,
}

word_index = {**special_tokens, **adjusted_word_index}
```

```python
def simple_encode(text):
    words = text.lower().split()
    encoded = [1]  # Start token
    for word in words:
        encoded.append(word_index.get(word, 2))  # Unknown if not found
    return pad_sequences([encoded], maxlen=50, padding='post')

# Your custom sentence
text = "this movie was good and inspiring"

# Encode and predict
encoded = simple_encode(text)
pred = model.predict(encoded)

print("sentiment score:", pred[0][0])
print("prediction:", "Positive 😊" if pred[0][0] >= 0.5 else "Negative 😎")
```
